```python
import xml.etree.ElementTree as ET
import os
import cv2
```

```python
def listdirs(folder):
    return [
        d for d in (os.path.join(folder, d1) for d1 in os.listdir(folder))
        if os.path.isdir(d)
    ]
```

```python
sampleXML = os.path.join(demoApp,'hierarchy_1.xml')
```

```python
tree = ET.parse(sampleXML)
```

```python
root = tree.getroot()
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
GUIClass = ['Button', 'Spinner', 'Chronometer', 'ImageView', 'RatingBar',
            'ProgressBar', 'VideoView', 'ToggleButton', 'RadioButton',
            'Switch', 'Checkbox', 'ImageButton', 'TextView', 'EditText',
            'SeekBar']

def traverse(node, feature):
    nodeClass = node.attrib['class'].split('.')[2]
    if nodeClass in GUIClass:
        feature.append({nodeClass: node.attrib['bounds']})
    for child in node:
        traverse(child, feature)
```

```python
feature = []
for child in root:
    traverse(child, feature)
```
